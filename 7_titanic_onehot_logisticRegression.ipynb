````python
import pandas as pd
````

````python
dfx=pd.get_dummies(df['sex'])

#해당 컬럼의 문자가 숫자로 (female-> 0, male-> 1) 로 바뀐다 본인꺼는 1 로 본인아닌것은 0으로 채워짐

dfx
````

판다스의 컨켓으로 df에 dfx를 붙히기~ concat([블럭])에 넣어, 행 size 맞으면 알아서 union으로 만들어줄게 

그래프 보면 행쪽에 추가되니까 axis =1 추가해주는것


````python
df= pd.concat([df,dfx], axis=1) 
df
````

- pclass로 바뀐것 확인 , 원리는 벡터에서 왔다. (단위백터, 단위행렬)

pd.get_dummies(df['pclass']) 

- y(=servived) = a(x=pclass) + 0 <br>
- y= 0.42x(x값에 따라 기울기가 달라진다, 연속형으로 생각해서 기울기가 확 바뀜)<br>
- y= 0*x1 + 0*x2 + 1*x3-> pclass의 3번째가 됨<br>
- y= 1*x1 + 0*x2 + 0*x3-> pclass의 1번째가 됨<br>

pd.get_dummies(df['pclass'], prefix='pclass', drop_first=True)
*   0       1        0 ->pclass_2
*   0       0        1 ->pclass_3 이 두개가 아니면 pclass_1이 되기 때문에 
* ( 1       0        0 ->pclass_1)삭제 한 것!

````python
dfx = pd.get_dummies(df['pclass'], prefix='pclass')
df = pd.concat([df,dfx], axis=1)
df.head(4)
````

후보군

female, male, pclass_1, pclass_2, pclass_3

````python
df[['fare', 'female', 'male', 'pclass_1', 'pclass_2', 'pclass_3', 'survived']].head(5)
````

연속형이 있을때는 표준정규분포롤 바꿔야 한다


````python
X=df[['fare', 'female', 'male', 'pclass_1', 'pclass_2', 'pclass_3']]
Y=df[['survived']]
X.shape, Y.shape
````


fit 교육시키고

규칙을 갖고있는 것에 데이터 넣어서 해당부분 뱉어내기





````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X.shape
````

split하기

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y)
````

LogisticRegresion 하기


````python
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
logR.fit(x_train, y_train)
````

````python
logR.score(x_train, y_train)
````
